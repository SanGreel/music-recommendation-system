```python
import numpy as np
import pandas as pd
import hashlib
import glob
import csv
```

```python
data_folder = "./data/audio/"
output = "./data/0_collection.csv"

files = glob.glob(data_folder + '/**/*.mp3', recursive=True)
```

```python
columns = ['i', 'id', 'name', 'filepath', 'genre']
df = pd.DataFrame(columns=columns)
```

```python
i = 0

for file in files:    
    fl_data = file.split("/")
    fl_data = fl_data[::-1]
    
    m_data = [[
            i,
            hashlib.md5(fl_data[0].encode('utf-8')).hexdigest()[0:10],
            fl_data[0],
            file,
            fl_data[1]
             ]]
    i+=1
    df_tmp = pd.DataFrame(m_data, columns=columns)
    df = df.append(df_tmp, ignore_index=True)
```

```python
df
```

```python
df.to_csv(output, sep='\t', encoding='utf-8')
```
